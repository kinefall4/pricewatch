# Community Price Watch – Automated Price Shock Reporting

## Problem (encountered in practice)
People experience price changes in everyday essentials, but tracking is usually manual and inconsistent (notes, WhatsApp messages, ad-hoc spreadsheets). This makes it hard to notice sudden price shocks quickly and to communicate them clearly.

## Goal
Automate the pipeline:

**data → validation → change detection (shock alerts) → publish a readable report**

## Target users
- Households and community groups
- NGOs / local initiatives
- Students learning data + cloud automation

## Success criteria
- **Automation:** a GitHub push triggers CodeBuild and redeploys the site to S3 (CI/CD).
- **Practical signal:** detect large price changes (≥ 20%) as “shock alerts.”
- **Reproducibility:** infrastructure is provisioned with Terraform.
- **Publishing:** results are publicly accessible (live URL + GitHub repo).
